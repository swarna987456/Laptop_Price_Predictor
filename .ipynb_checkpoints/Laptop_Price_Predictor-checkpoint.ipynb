Screen Resolution:
Display resolution is expressed using horizontal and vertical pixel counts. The most frequently used resolutions on laptop and 2-in-1 PCs nowadays are 1366-by-768 (also known as HD) and 1920-by-1080 (Full HD or 1080p). 1920-by-1080 is the most appropriate screen resolution for laptops, in our opinion.

```python
def func(x):
    temp = x.split()[-1]
    return temp.split('x')    
```

To split a column containing lists of two values into two separate columns, we can use the pandas apply method along with the pd.Series constructor.

```python
df[['Horizontal_pixel','vertical_pixel']] = df['ScreenResolution'].apply(func).apply(pd.Series)
```

Tuning the Parameters.

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
```

Ridge()

```python
Ridge().get_params()
```

```python
search_space = {"model": [Ridge()],
                "model__alpha": np.logspace(0, 4, 10)}
```

```python
step1= ColumnTransformer(
transformers = [('OHE', OneHotEncoder(sparse = False,drop ='first',handle_unknown='ignore'),[0,1,7,10,11]),('scale',StandardScaler(),[2,3,6,9])],
remainder = 'passthrough')

step2 = Ridge()
    
pipeline = Pipeline([
    ('preprocessing',step1),
    ('model',step2)
    ])

search = GridSearchCV(pipeline, search_space, n_jobs=2,scoring = 'r2')
search.fit(X_train,y_train)
print(search.best_params_)
print(search.best_score_)
```

```python
best_model = search.best_estimator_

best_model.fit(X_train,y_train)

y_pred = best_model.predict(X_test)

print('train score',r2_score(y_train,best_model.predict(X_train)))
print('test score',r2_score(y_test,y_pred))

```

> Parameters that can increase predictive power of the model 
1.  'max_features': 'auto'
2. n_estimators
3. min_samples_leaf': 
4. max_depth: Setting a lower value for max_depth can simplify the decision tree and make it more generalized, while a higher value allows the tree to grow deeper and potentially capture more complex relationships in the data. However, it increases the risk of overfitting if the depth becomes too large.

> Parameters that make it easier to train the model:
 There are a few attributes which have a direct impact on model training speed. 
1. "model__n_jobs":[-1]
2. "model__random_state":[24]
 

```python
search_space ={"model": [RandomForestRegressor()],
                "model__n_estimators": [100,200,300],
                "model__max_features": [0.75,0.50],
                "model__max_depth":[3, 5, 10,15,20],
                 "model__n_jobs":[-1],
                 "model__random_state":[24],
              "model__max_samples":[0.5],
              "model__min_samples_leaf":[3,5,7]}
```

```python
step1= ColumnTransformer(
transformers = [('OHE', OneHotEncoder(sparse = False,drop ='first',handle_unknown='ignore'),[0,1,7,10,11])],
remainder = 'passthrough')

step2 = RandomForestRegressor()
    
pipeline = Pipeline([
    ('preprocessing',step1),
    ('model',step2)
    ])

search = GridSearchCV(pipeline, search_space, n_jobs=2,scoring = 'r2')
search.fit(X_train,y_train)
print(search.best_params_)
print(search.best_score_)
```

```python
best_model = search.best_estimator_

best_model.fit(X_train,y_train)

y_pred = best_model.predict(X_test)

print('train score',r2_score(y_train,best_model.predict(X_train)))
print('test score',r2_score(y_test,y_pred))
```

```python
# Define a range of k values to test
k_values = list(range(1, 21))

# Initialize lists to store the mean squared errors
mse_train = []
mse_test = []

# Iterate over the k values
for k in k_values:
    
    # Create a KNN regressor with the current k value
    #knn = KNeighborsRegressor(n_neighbors=k)
    
    # Fit the model on the training data
    step1= ColumnTransformer(
    transformers = [('OHE', OneHotEncoder(sparse = False,drop ='first',handle_unknown='ignore'),[0,1,7,10,11]),('scale',StandardScaler(),[2,3,6,9])],
    remainder = 'passthrough')

    step2 = KNeighborsRegressor(n_neighbors=k)
    
    pipeline = Pipeline([
    ('preprocessing',step1),
    ('model',step2)
    ])
    pipeline.fit(X_train, y_train)
    
    # Predict the target variable for the training and testing data
    y_train_pred = pipeline.predict(X_train)
    y_test_pred = pipeline.predict(X_test)
    
    # Calculate the mean squared error for training and testing data
    mse_train.append(mean_squared_error(y_train, y_train_pred))
    mse_test.append(mean_squared_error(y_test, y_test_pred))

# Plot the mean squared errors
plt.plot(k_values, mse_train, label='Train MSE')
plt.plot(k_values, mse_test, label='Test MSE')
plt.xlabel('k')
plt.xticks(range(1, 21))
plt.ylabel('Mean Squared Error')
plt.title('KNN Regressor')
plt.legend()
plt.show()
```
